```python
# WORK - INTRO PROGRAMMING
# AUTHOR: BRUNO BARBOSA
# REVISER: DIEGO PINHEIRO
# AIM: MONTE CARLO SIMULATION WITH TWO DICE


# MODULES --------------------------------------------------------------------------------------------------------------------------------------------
import random
import matplotlib.pyplot as plt
```

```python

# lista de faces do dado
die_faces = list(range(1,21))

# lista de probabilidades do D1
die1_prob = [1/20] * 20

# lista de probabilidades do D1
distribution = [0] * 21


results = []
num_simulations = 10000
for i in range(num_simulations):

    # sorteiro do D1
    D1_list = random.choices(die_faces, weights=die1_prob)
    D1 = D1_list[0]

    # Determina a probabilidade da face 20 no D2 (baseado no resultado do D1)
    d2_face20_prob = 1/D1

    # Determina as probabilidades do segundo dado como um todo
    die2_prob = [(1-d2_face20_prob)/19] * 19
    die2_prob.append(d2_face20_prob)

    # sorteio do segundo dado utilizando o vetor de probabilidades alterado
    D2_list = random.choices(die_faces, weights=die2_prob)
    D2 = D2_list[0]

    # Resultado do lancamento organizado em uma tupla
    draw = (D1, D2)
    # print(f"(D1, D2) = {draw}")

    distribution[max(draw)] = distribution[max(draw)] + 1

    # Store the higher of the two rolls 
    results.append(max(draw))

print(f"results = {len(results)}")

```

```python
distribution
print(f"Probabilidade: {distribution[20]/num_simulations}")
```

```python
plt.hist(results, bins=range(0,22), alpha=0.7, color='green', edgecolor='black')
plt.title('Histogram with Custom Bins')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
