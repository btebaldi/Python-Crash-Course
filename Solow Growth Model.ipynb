```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters for the model
K0 = 100             # Initial capital stock
s = 0.2               # Savings rate
delta = 0.05          # Depreciation rate
alpha = 0.3           # Output elasticity of capital
delta_t = 0.1         # Time step
T = 250               # Total time of simulation
N = int(T / delta_t)  # Number of time steps

# Initialize arrays to store the data
time = np.linspace(0, T, N)
capital = np.zeros(N)
capital[0] = K0  # Set initial capital

# Define the production function
def production(K):
    return 1*(K ** alpha)

# Implement the Forward Euler method to update capital
for i in range(1, N):
    dK = (s * production(capital[i-1]) - delta * capital[i-1]) * delta_t
    capital[i] = capital[i-1] + dK

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(time, capital, label='Capital Over Time')
plt.title('Solow-Swan Model Simulation')
plt.xlabel('Time')
plt.ylabel('Capital Stock')
plt.legend()
plt.grid(True)
plt.show()
```
